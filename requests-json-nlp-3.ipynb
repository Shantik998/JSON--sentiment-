## 2 
. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python
 import json
import lyricsgenius
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

genius = lyricsgenius.Genius('IyIPIWdDwCH9N0rBB--JhckhuX-84D9n2qxl3GeYCIHNWcWDx3u4LanWK19SKY2o')

# Search for the song "Who Says" by the artist "Selena Gomez".
song = genius.search_song("Who Says", "Selena Gomez")

# Look up song.
if song is not None and song.lyrics is not None:
    # Print the lyrics of the song.
    print(song.lyrics)

    # Save the lyrics to a JSON file.
    lyrics_data = {
        "song_title": song.title,
        "artist": song.artist,
        "lyrics": song.lyrics
    }

    with open("who_says_lyrics.json", "w") as json_file:
        json.dump(lyrics_data, json_file, indent=4)

    print("Lyrics saved to 'who_says_lyrics.json'.")
else:
    print("Lyrics not found for the song.")

# Read the contents of the JSON file.
with open("who_says_lyrics.json", "r") as json_file:
    lyrics_data = json.load(json_file)

# Extract the lyrics from the loaded JSON data.
lyrics = lyrics_data["lyrics"]

# Perform sentiment analysis using spaCyTextBlob.
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")
doc = nlp(lyrics)

# Print the polarity score of the sentiment analysis.
polarity_score = doc._.polarity
print("Polarity score:", polarity_score)
```

## 4. 
Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics? Why or why not do you think that might be?  Answer the questions in either a comment in the code cell or a markdown cell under the code cell.

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

def get_polarity_score(file_name):
    # Read the contents of the file.
    with open(file_name, "r") as file:
        lyrics = file.read()

    # Perform sentiment analysis using spaCyTextBlob.
    nlp = spacy.load("en_core_web_sm")
    nlp.add_pipe("spacytextblob")
    doc = nlp(lyrics)

    # Get the polarity score of the sentiment analysis.
    polarity_score = doc._.polarity
    return polarity_score

# Get polarity scores for the three song files.
song_files = ["who_says_lyrics.json", "hymn_for_the_weekend_lyrics.json"]

for file in song_files:
    polarity_score = get_polarity_score(file)
    print(f"Polarity score for {file}: {polarity_score}")
```
